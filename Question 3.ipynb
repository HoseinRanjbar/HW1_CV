```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def myImageFilter(img0, h):
 
    img1=np.zeros_like(img0)
    w , t = np.shape(h)[0] , np.shape(h)[1]
    # Put pad for protect the imformations
    img0=np.pad(img0, ((w-1, w-1), (t-1,t-1)), 'edge')

    # Apply filter to the image
    for i in range(np.shape(img0)[0]-2*(w-1)):
        for j in range(np.shape(img0)[1]-2*(t-1)):
            img1[i,j]=np.sum(h*img0[i:i+w,j:j+t])
           
    plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))
    
    return img1
```

```python
# Test myImageFilter function with mean filter
img= cv2.imread('parrots.png',cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
h=np.zeros((20,20))
h.fill(1/400)
myImageFilter(img,h)
```
